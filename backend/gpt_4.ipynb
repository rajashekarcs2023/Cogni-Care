```python
import base64
import requests
import openai
import os
import json
```

```python
api_key = "APIKEY1"

def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
def get_info(base64_image):
  headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
  }

  payload = {
    "model": "gpt-4-vision-preview",
    "messages": [
      {
        "role": "user",
        "content": [
          {
            "type": "text",
            "text": "Describe this image."
          },
          {
            "type": "image_url",
            "image_url": {
              "url": f"data:image/jpeg;base64,{base64_image}"
            }
          }
        ]
      }
    ],
    "max_tokens": 300
  }
  return headers, payload
```

```python
for folder1 in os.listdir(os.getcwd() + "\\snips\\"):
    for folder2 in os.listdir(os.getcwd() + "\\snips\\" + folder1):
        os.mkdir(os.getcwd() + "\\vid_summary\\" + folder1 + "\\" + folder2 + "\\")
```

```python
for folder1 in os.listdir(os.getcwd() + "\\snips\\"):
    for folder2 in os.listdir(os.getcwd() + "\\snips\\" + folder1):
        for image in os.listdir(os.getcwd() + "\\snips\\" + folder1 + "\\" + folder2):
            image_path = os.getcwd() + "\\snips\\" + folder1 + "\\" + folder2 + "\\" + image
            headers, payload = get_info(encode_image(image_path))
            
            response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

            with open(os.getcwd() + "\\vid_summary\\" + folder1 + "\\" + folder2 + "\\" + image.replace(".jpg", "") + ".json", "w") as outfile:
                print(response.json())
                outfile.write(response.json()['choices'][0]['message']['content'])
```
